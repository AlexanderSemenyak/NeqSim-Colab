```
#@title  Python-based Oil & Gas Price Analysis
#@markdown Oil Price Dataset: https://www.eia.gov/dnav/pet/hist_xls/RBRTEd.xls
import pandas as pd
url="https://www.eia.gov/dnav/pet/hist_xls/RBRTEd.xls"

bent_oil_prices=pd.read_excel(url,sheet_name="Data 1", skiprows=2455, names=['Dates', 'Brent Price'])
import matplotlib.pyplot as plt
plt.plot( 'Dates', 'Brent Price', data=bent_oil_prices)
plt.xlabel('Date')
plt.ylabel('Brent Price [Dollars per Barrel]')
plt.rcParams['figure.dpi'] = 200
bent_oil_prices.tail()
```

```
#@title  Python-based Oil & Gas Price Analysis
#@markdown Natural Gas DataSet: https://www.eia.gov/dnav/ng/hist_xls/RNGWHHDd.xls
import pandas as pd
url="https://www.eia.gov/dnav/ng/hist_xls/RNGWHHDd.xls"

henryhub_gas_prices=pd.read_excel(url,sheet_name="Data 1", skiprows=2455, names=['Dates', 'Henry Hub Gas Price'])
import matplotlib.pyplot as plt
plt.plot( 'Dates', 'Henry Hub Gas Price', data=henryhub_gas_prices)
plt.xlabel('Date')
plt.ylabel('Henry Hub Gas Price [Dollars Per Million Btu]')
plt.rcParams['figure.dpi'] = 200
henryhub_gas_prices.tail()
```
